## Age Distribution

```python
df['age'] = df['age'].astype(float)

dia = sns.histplot(df, x= 'age')
for container in dia.containers:
    dia.bar_label(container)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

```python
sns.boxplot(df, x='age')
```

```python
sns.violinplot(df, x='age')
```

## BMI Distribution

```python
dia = sns.histplot(df, x='bmi')
for container in dia.containers:
    dia.bar_label(container)
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.show()
```

```python
sns.boxplot(df,x='bmi')
```

```python
sns.lineplot(df,x='age',y='bmi',hue='sex')
```

## Determine Gender data based on Smoking and Charges issued

```python
dia = sns.countplot(df,x='sex')
for container in dia.containers:
    dia.bar_label(container)
plt.title("Gender Count")
plt.xlabel("Gender")
plt.show()
```

```python
dia = sns.countplot(df,x='smoker',hue='sex')
for container in dia.containers:
    dia.bar_label(container)
plt.title("Smokers Count on Gender")
plt.xlabel("Whether Smoke")
plt.show()
```

```python
dia = sns.barplot(df, x='smoker', y='charges',errwidth=0)
for container in dia.containers:
    dia.bar_label(container)
plt.title("Smoker VS Chargers Plot")
plt.xlabel("Whether Smoke")
plt.ylabel("Chargers")
plt.show()
```

```python
dia = sns.barplot(df, x='smoker', y='charges',hue='sex',errwidth=0)
for container in dia.containers:
    dia.bar_label(container)
plt.title("Smoker VS Chargers Plot based on Gender")
plt.xlabel("Whether Smoke")
plt.ylabel("Chargers")
plt.show()
```

## Population distribution across Region

```python
dia = sns.countplot(df,x='region',hue='sex')
for container in dia.containers:
    dia.bar_label(container)
plt.title("Gender Distribution in Different Region")
plt.xlabel("Region")
plt.show()
```

## Count of Children

```python
dia = sns.countplot(df,x='children')
for container in dia.containers:
    dia.bar_label(container)
plt.title("Childeren Count")
plt.xlabel("Children")
plt.show()
```

## PairPlot based on Smokers count and Gender

```python
g = sns.pairplot(df, hue ='smoker')
g.figure.suptitle("Pairplot based on Smokers")
plt.show()
```

```python
g = sns.pairplot(df,hue='sex')
g.figure.suptitle("Pairplot based on Gender")
plt.show()
```

## Heatmap of all Numeric Values to define Co-relation

```python
corr = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True)
```
